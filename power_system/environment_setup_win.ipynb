#Python+ipython+pylab安装教程

- 下载安装Python


    - 下载: python-3.4.3.amd64.msi :[官方网站](https://www.python.org/downloads/)
    - 安装: 直接双击安装
    
- 设置环境变量 



    -  把c:\Python34 和c:\Python34\Scripts 都添加到环境变量中. 右击我的电脑——高级系统设置——环境变量——在系统变量里有path选项——双击打开——在原有变量的基础上添加英文状态下的分号——然后将路径名输入即可。（切记，不要删除原先的系统变量，只要用分号隔开，然后添加，最后也要加上分号）
    
    
- 下载安装IPython: 

 在cmd输入:
```
pip install ipython
```
 pip是获取python模块的最有效办法，程序运行的各种模块都推荐采用这种方式安装。
 
 接下来输入：
```
ipython
```
确认是否安装成功。


- notebook安装：
 ipython notebook,现在称为jupyter,它的理念十分前卫，是教学、计算、科研的一个重要工具，具体参见[IPython Notebook简介](http://hyry.dip.jp/tech/slice/slice.html/35).notebook可以保存你的研究过程，并和其他人共同研究。自从iPython4.0开始，notebook已经属于jupyter这个软件包了。
 

 
 先安装pylab，windows下需要到 *http://www.lfd.uci.edu/~gohlke/pythonlibs* 下载：
 
 [numpy](http://www.lfd.uci.edu/~gohlke/pythonlibs/#numpy)
 [scipy](http://www.lfd.uci.edu/~gohlke/pythonlibs/#scipy)
 [matplotlib](http://www.lfd.uci.edu/~gohlke/pythonlibs/#matplotlib)
 
 切换下载文件的目录，在cmd中用：
 
```
 pip install [文件名]
```
    进行安装。
    
 然后，安装jupyter:
```
 pip install "jupyter"
```
 
 执行完上述步骤后，恭喜你，你已经拥有了完整的python运行环境以及目前最先进的科研计算开发平台。  
 只要在命令行中键入：
```
 jupyter notebook
```

就可以启动notebook进程，并自动弹出浏览器界面。
